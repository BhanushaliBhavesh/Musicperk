```python
import pandas as pd
import matplotlib.pyplot as plt
import mysql.connector
import mplfinance as mpf
%matplotlib inline
```

```python
mydb=mysql.connector.connect(host='localhost',user='root',password='admin',database='musicperk')
```

```python
data = pd.read_sql_query('Select * from hindalco;', mydb)
```

```python
data = data.set_index('dat')
data.head()
```

```python
data.tail()
```

```python
data_open = pd.DataFrame(data.open)
```

```python
data_open['MA_44'] =data_open.open.rolling(44).mean()
```

```python
data['open'].plot(label = 'Hindalco Open Price',figsize=(15,7))
plt.legend()
plt.title('Open price')
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Open price and MA44')
plt.plot(data_open['open'],label='open')
plt.plot(data_open['MA_44'],label='MA44')
plt.legend()
plt.show()
```
